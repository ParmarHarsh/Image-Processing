```python
# call another jupyter notebook to run required function
%run 'MyCanny_script.ipynb'

# import libraries
import matplotlib.pyplot as plt
import kornia
import torch

# load the images
img1 = plt.imread('bowl-of-fruit.jpg')
img2 = plt.imread('desert-fox.jpg')

# convert images to PyTorch tensors and normalize
t_img1 = torch.from_numpy(img1.transpose(2,0,1)).float() / 255.0
t_img1 = t_img1.unsqueeze(0)
t_img2 = torch.from_numpy(img2.transpose(2,0,1)).float() / 255.0
t_img2 = t_img2.unsqueeze(0)

# convert images to grayscale
g_img1 = kornia.color.rgb_to_grayscale(t_img1)
g_img2 = kornia.color.rgb_to_grayscale(t_img2)

# convert tensors to numpy arrays
g_img1 = g_img1.squeeze(0).squeeze(0).numpy()
g_img2 = g_img2.squeeze(0).squeeze(0).numpy()

# apply canny edge detection
nms_img1, output_img1 = MyCanny(g_img1, st_deviation=5.0, gm_threshold=0.2)
nms_img2, output_img2 = MyCanny(g_img2, st_deviation=3.0, gm_threshold=0.2)

# display original and processed images
plt.imshow(img1, cmap='gray')
plt.title('Original Bowl of Fruit Image')
plt.axis('off')
plt.show()
plt.imshow(output_img1, cmap='gray')
plt.title('Canny Edge Detection on Bowl of Fruit Image')
plt.axis('off')
plt.show()

plt.imshow(img2, cmap='gray')
plt.title('Original Desert Fox Image')
plt.axis('off')
plt.show()
plt.imshow(output_img2, cmap='gray')
plt.title('Canny Edge Detection on Desert Fox Image')
plt.axis('off')
plt.show()
```
